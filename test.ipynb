```kotlin
import com.davidlukash.jsonmath.buildFunction
import com.davidlukash.jsonmath.createExpression
import kotlinx.serialization.json.Json

val json = Json {
    prettyPrint = true
    allowStructuredMapKeys = true
}

val call = buildFunction {
    name = "control.forEach"
    appendString("locals.name")
    appendList {
        appendString("Happy Face")
        appendString("Medium Face")
        appendString("Neutral Face")
        appendString("Neutral Sad Face")
        appendString("Sad Face")
        appendString("Oven")
        appendString("Ingredient Shop")
        appendString("Upgrade Shop")
        appendString("Missing")
    }
    appendList {
        appendFunction {
            name = "variable.set"
            appendString("globals.items")
            appendFunction {
                name = "list.join"
                appendFunction {
                    name = "variable.get"
                    appendString("globals.items")
                    appendBoolean(true)
                }
                appendFunction {
                    name = "list.construct"
                    appendFunction {
                        name = "dictionary.join"
                        appendFunction {
                            name = "dictionary.construct"
                            appendList {
                                appendString("name")
                            }
                            appendFunction {
                                name = "list.construct"
                                appendFunction {
                                    name = "variable.get"
                                    appendString("locals.name")
                                }
                            }
                        }
                        appendDictionary {
                            appendPair(
                                keyBuilder = {
                                    string("type")
                                },
                                valueBuilder = {
                                    string("INGREDIENT")
                                }
                            )
                            appendPair(
                                keyBuilder = {
                                    string("amount")
                                },
                                valueBuilder = {
                                    number("1000")
                                }
                            )
                            appendPair(
                                keyBuilder = {
                                    string("price")
                                },
                                valueBuilder = {
                                    number("0")
                                }
                            )
                            appendPair(
                                keyBuilder = {
                                    string("fastPriceGrowth")
                                },
                                valueBuilder = {
                                    boolean(false)
                                }
                            )
                            appendPair(
                                keyBuilder = {
                                    string("total")
                                },
                                valueBuilder = {
                                    number("0")
                                }
                            )
                            appendPair(
                                keyBuilder = {
                                    string("increment")
                                },
                                valueBuilder = {
                                    number("1000")
                                }
                            )
                            appendPair(
                                keyBuilder = {
                                    string("increaseSlope")
                                },
                                valueBuilder = {
                                    number("0")
                                }
                            )
                            appendPair(
                                keyBuilder = {
                                    string("cakePriceAccountability")
                                },
                                valueBuilder = {
                                    dictionary {
                                        appendPair(
                                            keyBuilder = {
                                                number("1")
                                            },
                                            valueBuilder = {
                                                number("0")
                                            }
                                        )
                                    }
                                }
                            )
                        }
                    }
                }
            }
            appendBoolean(true)
        }
    }
}

json.encodeToString(createExpression(call))
```
