```kotlin
import kotlin.random.Random
import kotlin.random.nextLong


fun weightedRandom(weight: Float, max: Float, random: Random): Float {
    val n = random.nextFloat()
    val weighted = n.pow(1f / weight)

    return weighted * max
}

fun weightedRandomInt(weight: Float, max: Int, random: Random): Int {
    return minOf(floor(weightedRandom(weight, max.toFloat(), random)).toInt(), max - 1)
}

fun <T> weightedRandomItem(weight: Float, items: List<T>, random: Random): T {
    return items[weightedRandomInt(weight, items.size, random)]
}

val weight = 0.5f
val random = Random(Random.nextLong())
val items = listOf(1, 2, 3)

val numbers = (0..10).toList().map { weightedRandomItem(weight, items, random) }

//println("Numbers: $numbers")
println("Mean: ${numbers.sum().toFloat() / numbers.size.toFloat()}")
println("Max: ${numbers.max()}")
println("Min: ${numbers.min()}")

numbers.distinct().sorted().forEach { n ->
    println("$n: ${numbers.count { it == n }.toFloat() / numbers.size.toFloat()}, ${numbers.count { it == n }}")
}
```
